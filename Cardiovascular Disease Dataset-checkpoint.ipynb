## Outliers

```python
Q3=df['ap_hi'].quantile(0.75)
Q1=df['ap_hi'].quantile(0.25)
IQR=Q3-Q1
lower_limit=Q1-(1.5*IQR)
upper_limit=Q3+(1.5*IQR)
print("Quartile 3",Q3)
print("Quartile 1",Q1)
print("Inter_Quartile_Range",IQR)
print("lower_lmit is",lower_limit)
print("upper_limit",upper_limit)

```

## Outlier Calculation

```python
outliers=df[(df['ap_hi']<lower_limit)|(df['ap_hi']>upper_limit)]
print(outliers)
```

```python
df['ap_lo'].unique()
```

```python
Q1_lo=df['ap_lo'].quantile(0.25)
Q3_lo=df['ap_lo'].quantile(0.75)
IQR_lo=Q3_lo-Q1_lo
lower_limit_lo=Q1_lo-(1.5*IQR_lo)
upper_limit_lo=Q3_lo+(1.5*IQR_lo)
print("Quartile 3",Q3_lo)
print("Quartile 1",Q1_lo)
print("Inter_Quartile_Range",IQR_lo)
print("lower_lmit is",lower_limit_lo)
print("upper_limit",upper_limit_lo)
```

```python
outliers_lo=df[(df['ap_lo']<lower_limit_lo)|(df['ap_lo']>upper_limit_lo)]
print(outliers_lo)
```

```python
df.shape
```

## Removed ap_hi outliers

    --Outliers in ap_hi and ap_lo were removed rather than replaced for the following reasons:
1. Mean/Median replacement would introduce artificial blood pressure values in a medical dataset, which could mislead the model into learning incorrect patterns.
2. IQR-based replacement was rejected as it would overwrite genuine hypertension readings (180–200 mmHg) which are clinically valid and critical for cardiovascular disease prediction.
3. Values such as -100, 11500, and 1000 are clear data entry errors with no medically meaningful replacement value — removal is the only valid treatment.
4. After removal, approximately 65,000+ records remain which is sufficient for robust model training with no significant data loss."

```python
df=df[(df['ap_hi']>=80) & (df['ap_hi']<=200)]
print("after ap_hi cleaning:",df.shape)
```

```python
print("ap_hi min:", df['ap_hi'].min())
print("ap_hi max:", df['ap_hi'].max())
```

        ##Outliers in ap_hi were removed using the & operator, retaining only rows where systolic blood pressure falls between 80 and 200 mmHg.

## Removed ap_lo outliers

```python
df=df[(df['ap_lo']>=60) & (df['ap_lo']<=120)]
print("after ap_lo cleaning:",df.shape)
```

```python
print("ap_lo min:", df['ap_lo'].min())
print("ap_lo max:", df['ap_lo'].max())
```

## Modelling & Evaluation

## Algo 1 -Logistic Regresssion

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score
l=LogisticRegression()
l.fit(X_train,y_train)

ypred_train=l.predict(X_train)
print("Train_accuracy:",accuracy_score(y_train,ypred_train))
print("cross_validation:",cross_val_score(l,X_train,y_train,cv=5,scoring="accuracy").mean())

ypred_test=l.predict(X_test)
print("test_accuracy:",accuracy_score(y_test,ypred_test))
```

## Model Selection

## --XGboost classifier was selected as the best modely
   ## Model Selection

After comparing 7 machine learning algorithms — Logistic Regression, KNN, 
Decision Tree, Random Forest, AdaBoost, Gradient Boosting, and XGBoost — 
XGBoost was selected as the final model based on the following reasons:

1. Highest and most consistent performance across train (74.52%), 
   CV (73.35%) and test accuracy (73.45%) with minimal variance.

2. No overfitting — unlike Random Forest which showed a 27% gap between 
   train (97.84%) and test accuracy (70.71%).

3. Balanced precision and recall across both classes with an overall 
   F1-score of 0.73 confirming reliable predictions for both disease 
   and non-disease cases.

4. XGBoost is computationally efficient, supports built-in regularization, 
   and is widely adopted in industry for tabular healthcare data.

5. Systolic blood pressure (ap_hi), age, and cholesterol were identified 
   as the strongest predictors  consistent with established 
   cardiovascular disease risk factors in medical literature.

## Conclusion

    The feature importance analysis confirms that systolic blood pressure (ap_hi) is the dominant predictor of cardiovascular disease at 61.4%, overwhelmingly outweighing all other features. This leads to the core finding of this project — patients with high systolic blood pressure are significantly more prone to cardiovascular disease. This is consistent with decades of medical research establishing hypertension as the leading modifiable risk factor for heart disease and stroke. For a public health program, this finding suggests that blood pressure screening should be the primary focus of early cardiovascular disease detection initiatives.
